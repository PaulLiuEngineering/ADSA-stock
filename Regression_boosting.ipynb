```python
import pandas as pd
import sklearn
import os
os.chdir('..')
df1 = pd.read_csv('datasets/nyse/OP_NYSE.csv')
df1.drop(['Unnamed: 0.1'], axis = 1)
print(df1)
industries = df1['GICS Sub Industry']
industries_split = []
for n in industries:
    if n not in industries_split:
        industries_split.append(n)
d = {}
os.chdir('datasets_by_industry')
for name in industries_split:
    d[name] = df1[df1['GICS Sub Industry'] == name]
    d[name].to_csv(name+'.csv',index = False)

df_airlines = d['Airlines']
print(df_airlines)
```
